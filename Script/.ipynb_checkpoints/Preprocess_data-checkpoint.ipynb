```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 对原始Excel表格进行处理

```python
df_original = pd.read_csv('剩草量表_0725.csv')
# 日期的处理（该操作非幂等，重复操作会重复地在日期头部添加"2017-"）
df_original['日期'].replace(to_replace=['^', '月', '日'], value=['2017-', '-', ''], inplace=True, regex=True)
df_original.head()
```

牛舍和牛只类型的统计

```python
print df_original['牛舍'].value_counts()
print df_original['牛只类型'].value_counts()
```

## 数据字段处理

填充缺失值、对异常值进行特殊标注

```python
df_original['减草量'].fillna(value=0, inplace=True)
df_original['减草量'].replace(to_replace='加', value='-', inplace=True, regex=True)
df_original['牛舍'].replace(to_replace=['南', '北'], value=['S', 'N'], inplace=True, regex=True)
df_original['实际头均采食量'].fillna(value=np.nan, inplace=True)
df_original['实际头均采食量'].replace(to_replace='#DIV/0!', value=np.nan, inplace=True)

#for field in ['牛头数', '报草量']:
#    df_original[field].replace(to_replace=0.0, value=np.nan, inplace=True)

# 我们只关注“高产牛”的采食量预测，其他牛只数据不用于训练和预测。
df_original.loc[df_original['牛只类型'] != '高产', ['牛头数', '报草量','剩草量','实际头均采食量','头均产奶量']] = np.nan

#print df_original[df_original['日期'] == '2017-7-1']

# 观察几个字段的值的分布
for field in ['牛头数', '报草量', '剩草量']:
    #print field + "取值统计：", df_original[field].value_counts()
    print '字段：', field
    plt.cla()
    df_original[field].plot.hist(bins=200)
    plt.show()
    print '================================'


# 异常值处理：对于牛头数过少、报草量过低或剩草量过高的数据样本进行特殊标注，不用于训练和预测。
df_original.loc[(df_original['剩草量'] > 1000) | (df_original['牛头数'] < 75) | (df_original['报草量'] < 2000), ['牛头数', '报草量','剩草量','实际头均采食量','头均产奶量']] = np.nan

# 重新计算实际头均采食量，以减小舍入误差。
df_original['实际头均采食量'] = (df_original['报草量'].astype(np.float64) \
                              - df_original['减草量'].astype(np.float64) \
                              - df_original['剩草量'].astype(np.float64)) \
                              / df_original['牛头数'].astype(np.float64)

df = df_original.copy()
df.head()
```

## 输出清洗后、未添加新特征的的数据文件

```python
df.to_csv('data_elt_0305_0725.csv', encoding="utf_8_sig")
```

## 引入温度、湿度数据

请参考“中国天气网”上该网页的格式：http://www.weather.com.cn/weather40dn/101121007.shtml

该网页上伞的符号边的数被标注为“降水概率”。我们姑且把它和湿度混为一谈。

将研究对象时段的文本拷贝至文本文件temp_rh_raw.txt中。

**注意！** 剩草量表中日期从3月5日开始，但气温数据需要从3月4日开始计算。因为3月5日对应的条目含义为：“报草量为3月4号中午，3月4号晚上，3月5号早上的总草量”。可以理解为，该数据条目是在3月5日统计得出的，但上报草量的时间其实是3月4日早上。这个时候3月4日、5日的气温都是可以（通过天气预报）知道的。

```python
temp_hi_list, temp_low_list, rh_list = [], [], []
with open('temp_rh_raw.txt') as fin:
    lines = fin.readlines()
    for k, line in enumerate(lines):
        if k % 3 == 1:
            vals = line.strip()[:-3].split('/ ')
            temp_hi_list.append(float(vals[0]))
            temp_low_list.append(float(vals[1]))
        elif k % 3 == 2:
            rh = float(line.strip()[:-1]) / 100.0
            rh_list.append(rh)
    
print 'length:', len(temp_hi_list), len(temp_low_list), len(rh_list)
```

### 计算THI指标

```python
thi_list = 0.81 * np.array(temp_hi_list) + (0.99 * np.array(temp_hi_list) - 14.3) * np.array(rh_list)
thi_list = thi_list[3:]
print '从3月4日天气开始计算，共有', len(thi_list), '条天气数据'
```

### 将THI数据和已有数据拼接

此处也可以采用别的特征提取方式，例如单独把温度、湿度作为特征等等。

```python
def concat_temprh(cs, df_cs):
    df_thi = pd.DataFrame(thi_list, columns=['THI'], index=df_cs.index)
    df_cs_new = pd.concat([df_cs, df_thi], axis=1)
    return df_cs_new
```

## 引入泌乳天数和泌乳期（胎次）数据

该部分引用的数据文件calving_data.csv是process_calving_data.ipynb文件生成的。该数据文件的数据源是银香伟业采用的牧场管理系统（以色列出品的系统：阿菲金 Afimilk）。获得系统访问权限后，进入“动物”——“事件报告”——“事件列表”，即可访问一段时期内，所有被系统记录的事件，包括组别变更事件。此外在系统中的“动物”——“动物报告”中可以仿照“泌乳牛”报告，根据需要（定制化报告所需包含的字段）构造出包含所有牛只在**当前日期**的泌乳天数、泌乳期（胎次）、组别数据的报告。

```python
df_calving = pd.read_csv('calving_data.csv')

def concat_calving(cs, df_cs):
    df_calving[cs].fillna(value="[nan, nan]", inplace=True)
    calving_data = df_calving[cs].values
    
    try:
        calving_days = [float(x[1:-1].split(', ')[0]) for x in calving_data]
        calving_indices = [float(x[1:-1].split(', ')[1]) for x in calving_data]
    except:
        raise Exception('Please check the values in calving_data.csv')
    
    df_calving_numeric = pd.DataFrame(np.array([calving_days, calving_indices]).T, columns=['calving_days', 'calving_indices'], index=df_cs.index)
    #print df_calving_numeric
    
    df_cs_new = pd.concat([df_cs, df_calving_numeric], axis=1)
    return df_cs_new
```

以牛舍为单位，分别拼接更多的特征，包括THI和泌乳期（胎次）、泌乳天数。

```python
df_output = pd.DataFrame([])

for cs in set(df['牛舍'].values):
    selected_fields = ['日期', '牛舍', '实际头均采食量', '头均产奶量']
    df_cs = df[df['牛舍'] == cs].loc[:, selected_fields]
    df_cs.columns = ['date', 'cowshed', 'avg_actual', 'avg_milk']
    
    df_cs_thi = concat_temprh(cs, df_cs)
    df_cs_thi_calving = concat_calving(cs, df_cs_thi)
    #print df_cs_thi_calving.head()
    
    df_output = pd.concat([df_output, df_cs_thi_calving], axis=0)

print '数据样本数量：', len(df_output)
print df_output.head()
df_output.to_csv('data_output_0305_0725.csv')
```

# 对数据进行观察和关联分析

绘制各牛舍的实际头均采食量曲线，观察趋势。

```python
def plot_cs(cs):
    #cs = 'B1-1N'
    actual_eat = df[df['牛舍'] == cs]['实际头均采食量']
    df_date = pd.Series(np.array(actual_eat), index=pd.date_range('20170305', periods=len(actual_eat))).astype(np.float64)
    #print np.array(df_date)

    df_date.plot(label=cs)
```

```python
#for cs in ['B1-5S']:
for cs in set(df['牛舍'].values):
    plot_cs(cs)

plt.legend()
plt.show()
```

对于每个牛舍，绘制曲线观察趋势，并对各特征的时间序列做关联分析。

```python
def correlation(cs, df_cs):
    print '================', cs, '================'
    df_cs.index = df_cs['date'].values
    del df_cs['date'] # 无特殊意义，就是为了print .head()时好看
    #print df_cs.head()
    
    selected_fields = ['avg_actual', 'avg_milk', 'THI']
    df_cor = df_cs.loc[:, selected_fields]
    print df_cor.corr()
    
    df_cor.plot(style=['-', '-', '--'])
    plt.title(cs)
    plt.show()
    
    
    
for cs in set(df['牛舍'].values):
    df_cs = df_output[df_output['cowshed'] == cs]
    correlation(cs, df_cs)
```
